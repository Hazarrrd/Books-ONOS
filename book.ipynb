```python
import numpy


import numpy

new_books = [] 
shelves_notes = ""
shelves_table = []
for book in books : 
    new_book = ""
    
    if book.isbn  : 
        new_book +=(book.isbn )
        
    else: 
         new_book += "None"
    new_book += ", "
    new_book +=(book.id)
    new_book += ", "
    new_book +=(book.title )
    new_book += ", "
    if book.country : 
        new_book +=(book.country)
    else: 
         new_book += "None"
    new_book += ", "
    if book.year : 
        new_book +=(book.year)
    else: 
        new_book += "None"
    
    new_book += ", " 
    if book.month :
        new_book +=(book.month)
    else: 
        new_book += "None"
    new_book += ", "
    new_book +=(book.rate )
    new_book += ", "
    
    if book.pages :
        new_book +=(book.pages)
    else: 
        new_book += "None"
    new_book += ", "
    new_book +=(str(book.popularity)) 
    
    new_book += ", "
    if book.shelves: 
        shelf = [] 
        shelf.append(book.id)
        for s in book.shelves: 
            shelves_notes += book.id + ", " + s + "\n"
            shelf.append(s)
        new_book+=(book.shelves[0])
        shelves_table.append(shelf)
    else : 
        new_book+= "None"
        
    new_book += ", "
    new_book +=(book.authors[0])
    new_books.append(new_book)

books_list = ""

for book in new_books:
    books_list += book + "\n"
    
```
